```python
#step 1
##convert stop from numerical to string 
df['Stops'] = df['Stops'].apply(lambda x: 'Non-Stop' if x == 0 else 'Stop(s)')
df.head()
```

```python
#step 2
df = df.dropna() # drops any rows with missing values
df = df[df['Equipment'] != '\\N'] # removes any rows with invalid equipment values
```

```python
#step 3
##Find and drop Duplicates
df = df.drop_duplicates()
df.info()

#There are no duplicates
```

```python
##step 4
##convert all values in the variables to uppercase to ensure consistency.
df['Airline'] = df['Airline'].str.upper()
df['SourceAirport'] = df['Source airport'].str.upper()
df['DestAirport'] = df['Destination airport'].str.upper()
df['Equipment'] = df['Equipment'].str.upper()
df.head()
```

```python
##step 5
##
from fuzzywuzzy import fuzz

def fuzzy_match(source, target):
    ratio = fuzz.ratio(source.lower(), target.lower())
    if ratio >= 80:
        return target
    else:
        return source

df['SourceAirport'] = df.apply(lambda x: fuzzy_match(x['SourceAirport'], 'LAX'), axis=1)
df['DestinationAirport'] = df.apply(lambda x: fuzzy_match(x['DestinationAirport'], 'JFK'), axis=1)
df.head()
```
