```python
import pandas as pd
```

```python
#Read data
df = pd.read_csv("C:/Users/HP PC/Desktop/Preply/Analya/US_Accidents_Dec21_updated.csv")
df.head()
```

```python
#structure of the data
df.info()
```

```python
#1 drop non useful values
df.drop(['ID', 'End_Time', 'End_Lat', 'End_Lng', 'Description', 'Number', 'Street', 'Side', 'County', 'Zipcode', 'Country', 'Timezone', 'Airport_Code', 'Weather_Timestamp'], axis=1, inplace=True)

# Print the updated dataset
print(df.head())
```

```python
#2 data extractuion/selection steps

# convert relevant columns to datetime data type
df['Start_Time'] = pd.to_datetime(df['Start_Time'])

# create new columns to extract useful information
df['Year'] = df['Start_Time'].dt.year
df['Month'] = df['Start_Time'].dt.month
df['Day'] = df['Start_Time'].dt.day
df['Hour'] = df['Start_Time'].dt.hour
df['Weekday'] = df['Start_Time'].dt.weekday

# select relevant rows based on certain criteria
df = df.loc[(df['Severity'] >= 2) & (df['Visibility(mi)'] > 0)]

df.head()
```

```python
#Transform features

#check for missing values
df['Severity'].isna().value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

weather_encoder = LabelEncoder()
df['Weather_Condition'] = weather_encoder.fit_transform(df['Weather_Condition'])
df.head()
```

```python
##Engineer new feature
#df['Weekday'] = df['Start_Time'].dt.weekday < 5
df['Rush_Hour'] = df['Weekday'] & (((df['Hour'] >= 7) & (df['Hour'] <= 10)) | ((df['Hour'] >= 16) & (df['Hour'] <= 19)) | ((df['Hour'] >= 12) & (df['Hour'] <= 15) & (~df['Weekday'])))
df.head()
```

```python
##deal with missing data 

# Check missing values in each column
print(df.isnull().sum())

```

```python
# Replace missing values in Temperature(F) with the mean temperature
df['Temperature(F)'].fillna(df['Temperature(F)'].mean(), inplace=True)

# Replace missing values in Pressure(in) with the median pressure
df['Pressure(in)'].fillna(df['Pressure(in)'].median(), inplace=True)

# Replace missing values in Wind_Speed(mph) with the mode wind speed
df['Wind_Speed(mph)'].fillna(df['Wind_Speed(mph)'].mode()[0], inplace=True)

# Replace missing values in Weather_Condition with 'Unknown'
df['Weather_Condition'].fillna('Unknown', inplace=True)

# Drop rows with missing values in Start_Time and End_Time
df.dropna(subset=['Start_Time', 'End_Time'], inplace=True)
```

```python
dummy_vars = ['Severity', 'Wind_Direction', 'Traffic_Signal','Weather_Condition', 'Sunrise_Sunset', 'Civil_Twilight', 'Nautical_Twilight', 'Astronomical_Twilight']
df = pd.get_dummies(df, columns=dummy_vars, drop_first=True)
df.head()
```
